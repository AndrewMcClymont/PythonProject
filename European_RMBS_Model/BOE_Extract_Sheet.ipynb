```python
#Mapping the data for repayment
from src.utils.boe_extract_mapping import load_boe_mapping
from src.utils.boe_extract_mapping import apply_boe_mapping

mappings = load_boe_mapping("/Users/andrewmcclymont/PycharmProjects/PythonProject/European_RMBS_Model/Config/Mappings.yaml")

# Define which DataFrame column uses which YAML mapping
column_mapping = {
    "repayment_method": "repayment_method",
    "interest_rate_type": "interest_type",
    "current_interest_rate_index": "index_mapping"
}
# Apply mapping
boe_tape = apply_boe_mapping(df=boe_tape,column_mapping=column_mapping,mappings=mappings,default_value=999)
boe_tape.head()
```

```python
# Updating the logic to overwrite the part and part with grace period logic:
mask = (boe_tape['repayment_method_mapped'] == "part_and_part") & (boe_tape['principal_grace_period'] == "ND")
boe_tape.loc[mask, 'repayment_method_mapped'] = "repayment"

# Updating the logic to other with svr
mask = (boe_tape['current_interest_rate_index_mapped'] == "other")
boe_tape.loc[mask, 'current_interest_rate_index_mapped'] = "SVR"
```

```python
#Calculate the implied SVR
def implied_svr(df, interest_type_col, curr_interest_rate_col, interest_rate_margin_col):
    # Use np.where to vectorize the logic
    return np.where(
        df[interest_type_col] == "floating",
        df[curr_interest_rate_col] - df[interest_rate_margin_col],
        np.nan
    )


boe_tape['implied_svr'] = implied_svr(boe_tape,interest_type_col='interest_rate_type_mapped',curr_interest_rate_col="current_interest_rate",interest_rate_margin_col="current_interest_rate_margin")
print(boe_tape['implied_svr'])
```

```python
#Need to add some columns
from src.utils.date_functions import parse_date_columns_from_string

#Quarter (stored as Q3-xx)
boe_tape['maturity_quarter'] = boe_tape['date_of_loan_maturity'].str[:2]

#update all the dataframe for dates:
boe_tape = parse_date_columns_from_string(boe_tape, columns=['date_of_loan_maturity','pool_cut_off_date','interest_revision_date_1'], region="UK", day_of_month="end",change_all_day_of_month=False, suffix="_date_converted")

# pool_cut_off = boe_tape['pool_cut_off_date_date_converted'].drop_duplicates()
boe_tape['remaining_term'] = ((boe_tape['date_of_loan_maturity_date_converted'].dt.year - boe_tape['pool_cut_off_date_date_converted'].dt.year) * 12
    + (boe_tape['date_of_loan_maturity_date_converted'].dt.month - boe_tape['pool_cut_off_date_date_converted'].dt.month))


boe_tape['reversion_term'] = ((boe_tape['date_of_loan_maturity_date_converted'].dt.year - boe_tape['pool_cut_off_date_date_converted'].dt.year) * 12
    + (boe_tape['date_of_loan_maturity_date_converted'].dt.month - boe_tape['pool_cut_off_date_date_converted'].dt.month))


boe_tape['reversion_term'] = ((boe_tape['interest_revision_date_1_date_converted'].dt.year - boe_tape['pool_cut_off_date_date_converted'].dt.year) * 12+ (boe_tape['interest_revision_date_1_date_converted'].dt.month - boe_tape['pool_cut_off_date_date_converted'].dt.month))

boe_tape.head()
```
